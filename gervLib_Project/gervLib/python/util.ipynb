```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import cm
import matplotlib.patches as mpatches
import io
import seaborn as sns
import os
from os import listdir
from os.path import isfile, join
%matplotlib inline
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
plt.rcParams["axes.titlesize"] = 12 
plt.rcParams["figure.figsize"] = (3,3)
```

```python
path = "../datasets/Dataset1.csv"
df = pd.read_csv(path, sep=' ', header=None)
df.columns = ["D1", "D2"]
df["type"] = "Instância"
custom_palette = {
        'Instância': 'tab:blue',
        'Pivô': 'tab:red',
        'Auxiliar': 'tab:green'
}
```

```python
df.at[18, "type"] = "Pivô"
circle1 = plt.Circle(xy=(0, 0), radius=11.67, color='black', fill=False)


p1 = sns.scatterplot(data=df, x="D1", y="D2", hue="type", palette=custom_palette)
p1.set(xticks=np.arange(0,21,10))
p1.set_xticklabels(np.arange(0,21,10))
p1.set(yticks=np.arange(0,21,10))
p1.set_yticklabels(np.arange(0,21,10))
p1.add_patch(circle1)
p1.set_aspect('equal')
p1.legend().set_visible(False)
p1.set(xlabel=None)
p1.set(ylabel=None)
plt.savefig("lvl1_vp.svg", dpi=300)
```
